```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CC GENERAL.csv')
```

```python
df = df.drop('CUST_ID',axis= 'columns')
```

```python
df = df.dropna(subset=['MINIMUM_PAYMENTS'])
```

```python
df = df.dropna(subset=['CREDIT_LIMIT'])
```

```python
df.describe()
```

```python
df.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df,orient='h')
plt.title('Box Plot for every variable')
plt.show()
```

```python
outliers_cols = [
    'BALANCE',
    'BALANCE_FREQUENCY',
    'PURCHASES',
    'ONEOFF_PURCHASES',
    'INSTALLMENTS_PURCHASES',
    'CASH_ADVANCE',
    'PURCHASES_FREQUENCY',
    'ONEOFF_PURCHASES_FREQUENCY',
    'PURCHASES_INSTALLMENTS_FREQUENCY',
    'CASH_ADVANCE_FREQUENCY',
    'CASH_ADVANCE_TRX',
    'PURCHASES_TRX',
    'CREDIT_LIMIT',
    'PAYMENTS',
    'MINIMUM_PAYMENTS',
    'PRC_FULL_PAYMENT',
    'TENURE'
]

for column in outliers_cols:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df = df[
            (df[column] >= lower_bound) & (df[column] <= upper_bound)
        ]
```

```python
plt.figure(figsize=(20, 20))
sns.boxplot(data=df,orient='h')
plt.title('Box Plot for every variable')
plt.show()

# for col in outliers_cols:
#     sns.boxplot(df[col])
#     plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
X = df[outliers_cols]
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++')
kmeans.fit(X)
y_kmeans = kmeans.fit_predict(X.values)
```

```python
df.head()
```

```python
plt.scatter(X.values[:, 0], X.values[:, 1], c=y_kmeans, s=50, cmap='viridis')
```

```python
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```
