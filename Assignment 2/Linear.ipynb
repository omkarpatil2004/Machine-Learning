```python
x = df[['Area']].astype(float).values 
y = df[['Price']].astype(float).values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
y_pred = lr.fit(x,y)
```

```python
y_pred.predict(x)
```

```python
y_prediction = y_pred.predict(x)
```

```python
print(y_pred.coef_) #this is m value 'Slope'
```

```python
print(y_pred.intercept_) #this is b value 'Intercept'
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y,y_prediction) #MSE
```

```python
r2_score(y,y_prediction) #R2 Score
```

```python
#gradient descent
epochs = 15

def gradient(x,y):
    m = 0
    b = 0
    n = 10  
    lr = 0.1

    
    for i in range(epochs):
        y_pred = m * x + b  
        cost = (1/n) * sum(y - y_pred)** 2  
        
        dm = (-2/n) * sum(x * (y - y_pred))  
        db = (-2/n) * sum(y_pred - y) 
        
        m = m - lr * dm  
        b = b - lr * db  
        
        print(cost,m,b)

gradient(x,y)
```
